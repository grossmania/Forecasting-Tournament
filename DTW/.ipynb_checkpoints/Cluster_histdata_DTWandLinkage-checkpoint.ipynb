# Identifying the similarity of time series data
##  Using dynamic time wraping and clustering

The following code produces the results as presented in the related word file. Additional analysis to test robustness and functionality are presented in the appendix.

```python
### Houskeeping / Packages

from dtaidistance import dtw
from dtaidistance import clustering
from dtaidistance import dtw_visualisation as dtwvis
from sklearn import preprocessing

import numpy as np
import pandas as pd
import math

import matplotlib.pyplot as plt

import scipy.cluster.hierarchy as sch
from scipy.spatial.distance import squareform

%matplotlib inline
```

### Preprocessing

```python
### Import (hist) data from csv file

# read csv-file
df_hist = pd.read_csv('../historical_data.csv')

# change column names
df_hist.rename(columns={'Unnamed: 0': 'month'}, inplace=True)
df_hist.rename(columns={'Month': 'month_bw'}, inplace=True)

# replace "nan"-cells with 0
df_hist.fillna(0, inplace=True)

# df_hist.head()
```

```python
### Scaling of the time series data
# With MinMaxScaler() 
# With (x-xmin)/(xmax-xmin) for each column (only range of data changes between 0 and 1)

# Scale dataframe
scaler = preprocessing.MinMaxScaler() 
d = scaler.fit_transform(df_hist)
df_hist_scale = pd.DataFrame(d, columns=df_hist.columns)

# Drop first to columns (representing row names) for easier handling of next steps
df_hist_scale = df_hist_scale.drop(["month", "month_bw", "COVID_deaths", "COVID_cases"], axis=1)

# df_hist_scale.head()
```

### Dynamic Time Wraping

```python
### Create distance matrix with pairwise comparison

# Transpose to array that observations (each time series) are in rows
array_scale = np.transpose(np.array(df_hist_scale))

# Calculate distrance matrix
distance_matrix_scale = pd.DataFrame(dtw.distance_matrix(array_scale, window=12),
                               columns=df_hist_scale.columns,
                               index=df_hist_scale.columns)

distance_matrix_scale.head()
```

### Clustering

```python
### Clustering (using the distance matrix)

# SciPy linkage clustering MinMaxScale()-based

model_scale = clustering.LinkageTree(dtw.distance_matrix, {})
cluster_idx_scale = model_scale.fit(array_scale)
```

```python
### Plotting Dendrogram based on MinMaxScaler()

# LinkageTree MinMaxScaler()-based
fig_scale, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 10))
show_ts_label = df_hist_scale.columns
model_scale.plot("LinkageTree_Dendrogram.png", axes=(ax[0],ax[1]), ts_height=20,
            bottom_margin=5, top_margin=0.1,
            show_ts_label=show_ts_label, ts_label_margin=-10,
            show_tr_label=True, tr_label_margin=0.1,
            ts_left_margin=10, ts_sample_length=5,
            tr_left_margin=1,
            cmap='viridis_r', ts_color=None)

fig_scale.suptitle('Dendrogram timeseries cluster based on MinMaxScaler()', fontsize=16)
#fig_1.tight_layout()
fig_scale
```

```python
### Select cut-off value

cutoff = 1.44
#cutoff = input ("Enter cutoff value based on dendrogram structure :")

cluster_assignment = sch.fcluster(model_scale.linkage, t=cutoff, criterion="distance")
```

```python
### Save cluster results in table

df_cluster_result = pd.DataFrame(np.zeros((1,len(cluster_assignment))), 
                         columns=df_hist_scale.columns, 
                         index=['cluster'])

for j in range(0,len(df_cluster_result.columns),1):
    df_cluster_result.iloc[0,j] = cluster_assignment[j]

df_cluster_result
```

### Calculate statistical time series properties

```python
temp = pd.DataFrame(np.zeros((3,len(df_cluster_result.columns))), 
                    columns=df_hist_scale.columns, 
                    index=['curve_length','var','std_dev'])

df_stat_result = df_cluster_result.append(temp)

# df_stat_result
```

```python
#Calculate curve length, variance and standard devidation for each timeseries

for j in range(0,len(df_stat_result.columns),1):
    
    curve_length = 0
    y = np.array(df_hist_scale.iloc[:,j])
    y = np.insert(y, 0, 0)
    
    for i in range(0,y.size-2,1):
        curve_length = curve_length + math.sqrt(1+(y[i+1]-y[0])**2)
    
    #df_tsprop.iloc[0,j] = cluster_assignment[j]
    df_stat_result.iloc[1,j] = curve_length
    df_stat_result.iloc[2,j] = np.var(df_hist_scale.iloc[:,j])
    df_stat_result.iloc[3,j] = np.std(df_hist_scale.iloc[:,j])
    
df_stat_result.round(3)
df_stat_result.to_excel('result_table_cluster.xlsx')
```

### Testing dtaidistance-package by using its reference packages directly

I use the scipy library directly to see which input the cluster function uses and if the default methods are adopted in the dtaidistance package.

1. Input must be a condensed matrix. Hence, dtaidistance.clustering.LinkageTree internally transforms the input matrix (not condensed) into a condensed matrix.
2. "complete" method is used to calculate cluster distance

```python
import scipy.cluster.hierarchy as sch
from scipy.spatial.distance import squareform

condensed_distance_matrix = squareform(dtw.distance_matrix(array_scale)) #add window here if needed
# --> condensed distance matrix is used                                  #remember to also change cutoff

##Possible cluster distance methods
#single
#complete
#average
#weighted
#centroid
#median
#ward

Z = sch.linkage(condensed_distance_matrix, method='complete')
# --> "complete" method is used to calculate cluster distance

fig = plt.figure(figsize=(25, 10))

dn = sch.dendrogram(Z)
# --> same dendrogram as in main results above

### Save the cluster results
cluster_assignment_check = sch.fcluster(Z, t=cutoff, criterion="distance")

df_cluster_result_check = pd.DataFrame(np.zeros((1,len(cluster_assignment_check))), 
                         columns=df_hist_scale.columns, 
                         index=['cluster'])

for j in range(0,len(df_cluster_result_check.columns),1):
    df_cluster_result_check.iloc[0,j] = cluster_assignment_check[j]
```
